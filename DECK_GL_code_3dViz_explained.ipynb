```
def in_bounding_box(lng, lat):
    return (DOWNTOWN_BOUNDING_BOX[0] <= lng <= DOWNTOWN_BOUNDING_BOX[2] and
            DOWNTOWN_BOUNDING_BOX[1] <= lat <= DOWNTOWN_BOUNDING_BOX[3])

# Filter
df = df[df.apply(lambda row: in_bounding_box(row["lng_w"], row["lat_w"]), axis=1)
        & df.apply(lambda row: in_bounding_box(row["lng_h"], row["lat_h"]), axis=1)]

# Define source and target colors
GREEN = [0, 255, 0, 40]
RED = [240, 100, 0, 40]

# Create the ArcLayer in pydeck
arc_layer = pdk.Layer(
    "ArcLayer",
    data=df,
    get_source_position=["lng_h", "lat_h"],   # home (source) position
    get_target_position=["lng_w", "lat_w"],   # work (destination) position
    get_source_color=RED,
    get_target_color=GREEN,
    get_width="S000 * 2",   # width accessor: uses a column named "S000" from the dataset
    get_tilt=15,
    pickable=True,
    auto_highlight=True,
)

# Define an initial view state (center, zoom, pitch, etc.)
view_state = pdk.ViewState(
    latitude=37.7576171,
    longitude=-122.5776844,
    zoom=8,
    bearing=45,
    pitch=50
)

# Tooltip template (to show on hover)
tooltip = {
    "html": "{S000} jobs <br> Home location (red) → Work location (green)",
    "style": {"color": "white"},
}

# Create the Deck object
deck = pdk.Deck(
    layers=[arc_layer],
    initial_view_state=view_state,
    tooltip=tooltip,
)

# Display in notebook / Colab
deck.show()

# (Optional) Save to HTML file if you want to export
deck.to_html("arc_layer_map.html", notebook_display=True)
```
