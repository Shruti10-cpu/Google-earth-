```python
import pydeck as pdk
import pandas as pd
```

```python
DATA_URL ="https://raw.githubusercontent.com/ajduberstein/geo_datasets/master/small_waterfall.csv"
```

```python
df = pd.read_csv(DATA_URL)
df.head()
```

```python
point_cloud = pdk.Layer(
    "PointCloudLayer",
    data=df,
    get_position=["x", "y", "z"],
    get_color=["r", "g", "b"],
    get_normal=[0, 0, 15],
    point_size=2,
    pickable=True,
    auto_highlight=True,
)

view = pdk.ViewState(
    target=[df.x.mean(), df.y.mean(), df.z.mean()],
    rotation_x=15,
    rotation_orbit=30,
    zoom=5,
    controller=True,
)

view3d = pdk.View(type="OrbitView", controller=True)
deck = pdk.Deck(layers=[point_cloud], initial_view_state=view, views=[view3d])
deck.show()
dewck.to_html("point_cloud.html")
```
